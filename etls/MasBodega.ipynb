```python
sem = pd.read_excel("/content/drive/MyDrive/Sell Out/Mas Bodega/Vta/Mensual/2024/12_24.xlsx")

#sem = pd.read_excel("/content/drive/MyDrive/Sell Out/Mas Bodega/Vta/Semanal/2024/45_24.xlsx")
sem.head()
```

Ajustar columnas

```python
sem = sem[1:]  #eliminar las primeras dos filas

sem = sem.iloc[:,[0,1,2,3,4,5]]  #seleccionar columnas por index

#Renaming columns by index (el nuevo dataset tiene nuevas columnas ordenadas)
sem.columns.values[0] = "prod"
sem.columns.values[1] = "descripcion"
sem.columns.values[2] = "determinante"
sem.columns.values[3] = "nom tda"
sem.columns.values[4] = "cjas"
sem.columns.values[5] = "importe"

sem.head()
```

Revisar valores del dataset

```python
sem.info()
```

```python
print("Missing values distribution: ")
print(sem.isnull().mean())
print("")
sem.shape
```

Transformar el tipo de datos del dataset

```python
sem["cjas"] = sem["cjas"].astype(float)
sem["importe"] = sem["importe"].astype(float)
```

Valores negativos a positivos

```python
sem['cjas'] = np.where(sem['cjas'] < 0, 0, sem['cjas'])
sem['importe'] = np.where(sem['importe'] < 0, 0, sem['importe'])
```

seleccionar valores mayores a 0´s

```python
sem = sem[(sem['cjas']>0) & (sem['importe']>0)]
```

Eliminar todo los espacios en blanco del dataset final

```python
sem = sem.replace(r'\s+', ' ', regex=True) # espacios en blanco
```

# Inventarios

```python
stocks = pd.read_excel("/content/drive/MyDrive/Sell Out/Mas Bodega/Invt/Mensual/2024/12.xlsx")

#stocks = pd.read_excel("/content/drive/MyDrive/Sell Out/Mas Bodega/Invt/Semanal/2024/45_24.xlsx")

stocks=stocks.replace(r'\s+', ' ', regex=True)

stocks.head()
```

Delimitar el dataset

```python
stocks = stocks[1:]  #eliminar las primeras dos filas

stocks = stocks.iloc[:, [0,1,2,3,5]]  #seleccionar columnas por index

#Renaming columns by index (el nuevo dataset tiene nuevas columnas ordenadas)
stocks.columns.values[0] = "determinante"
stocks.columns.values[1] = "nom tda"
stocks.columns.values[2] = "prod"
stocks.columns.values[3] = "descripcion"
stocks.columns.values[4] = "invt cjas"

stocks.head()
```

```python
stocks.info()
```

```python
print("Missing values distribution: ")
print(stocks.isnull().mean())
print("")
stocks.shape
```

Transformar el tipo de dato

```python
stocks["invt cjas"] = stocks["invt cjas"].astype(float)
```

valores negativos

```python
stocks['invt cjas'] = np.where(stocks['invt cjas'] < 0, 0, stocks['invt cjas'])
```

valores nulos

```python
stocks = stocks[(stocks['invt cjas']>0)]
```

### vtas + invt

```python
sos=pd.concat([sem,stocks])  #unir verticalmente los datsets

sos.fillna(0, inplace=True)  #rellenar los espacios con 0
```

ELIMINAR los siguientes centros de venta

```python
#eliminar los siguientes puntos de venta
condition =(sos['determinante'] == 20)

# Delete rows based on the condition
sos = sos[~condition]
```

# Merged catalogos

```python
#productos
merged_sos= sos.merge(ab, left_on="prod", right_on="CLAVE", how="left")
```

```python
#tda
merged_sos = merged_sos.merge(tda, left_on="determinante", right_on="id", how="left")
```

### Pivot Table

```python
#tabla dinamica
pivot_table=pd.pivot_table(data=merged_sos,
                           index=["id", "AB actual"],
                           values=["cjas", "importe", "invt cjas"],
                           aggfunc=np.sum,
                           fill_value=0)

pivot_table.reset_index().to_excel('sos.xlsx',sheet_name = 'Sheet1',index=False)
```
