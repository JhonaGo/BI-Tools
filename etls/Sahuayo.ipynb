Carga del archivo

```python
sem = pd.read_excel("/content/drive/MyDrive/Sell Out/Sahuayo/vtas/Mensual/2025/01.25.xlsx", engine="openpyxl") # Explicitly use the openpyxl engine
sem.head()
```

Ajustes en el formato

```python
sem.columns = sem.iloc[1]  #indicar la primera columna que se encuentra en la fila index 1
sem = sem[2:]  #eliminar las primeras dos filas
sem.head()
```

Seleccion de columnas de interes

```python
sem.columns
```

```python
#select columns by index position
sem = sem.iloc[:, [0, 4,5,6,7,10]]
```

```python
sem.info()
```

Transformar el tipo de dato a considerar

```python
sem["Suma de Sell Out Cajas"] = sem["Suma de Sell Out Cajas"].astype(float)
sem["Suma de Sell Out Importe"] = sem["Suma de Sell Out Importe"].astype(float)

sem['Fecha'] = sem['Fecha'].astype('datetime64[ns]')
```

Seleccionar el rango de fechas

```python
#rango de fechas
start_date = datetime.strptime('11/04/2024', '%m/%d/%Y')
end_date = datetime.strptime('11/10/2024', '%m/%d/%Y')

#filtrar el rango de fechas dadas
sem = sem[(sem['Fecha'] >= start_date) & (sem['Fecha'] <=end_date)]
```

Valores perdidos en el dataset

```python
print("Missing values distribution: ")
print(sem.isnull().mean())
print("")
sem.shape
```

Renombrar columnas

```python
sem = sem.rename(columns={'Suma de Sell Out Cajas': 'cjas', 'Suma de Sell Out Importe': 'importe'})
```

Sustituir valores negativos

```python
sem['cjas'] = np.where(sem['cjas'] < 0, 0, sem['cjas'])
sem['importe'] = np.where(sem['importe'] < 0, 0, sem['importe'])
```

seleccionar valores donde ambas columnas son mayores a 0 (vtas y $)

```python
#sem = sem[(sem['cjas']>0) & (sem['importe']>0)]
```

eliminar todo espacio en blanco una vez obtenido el dataframe final

```python
sem = sem.replace(r'\s+', ' ', regex=True) # espacios en blanco
sem = sem.drop ('Fecha', axis = 1) #eliminar fecha del dataset
sem.head()
```

## Merge catalogos

#### inventario

```python
#productos
merged_stocks = stocks.merge(ab, left_on="Clave Sic", right_on="CLAVE", how="left")
```

```python
#tiendas
merged_stocks = merged_stocks.merge(tda, left_on="Sucursal", right_on="Nombre de Tienda", how="left")
```

Datos faltantes en el dataset final

```python
print(merged_stocks.isnull().sum())
```

seleccionar registros con valores faltantes

## Pivot Table, semanal

```python
sos = pd.concat([merged_sem, merged_stocks])  #concatenar los datasets

pivot_table=pd.pivot_table(data=sos,
                           index=["id", "AB Nuevo"],
                           values=["cjas", "importe", "Suma de Cajas Inventario"],
                           aggfunc=np.sum,
                           fill_value=0)    #tabla dinamica

#exportar tabla dinamica
pivot_table.reset_index().to_excel('sos.xlsx',sheet_name = 'Sheet1',index=False)
```
