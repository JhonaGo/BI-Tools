#Merged Catalogos

###Venta

```python
#productos
merged_sem = sem.merge(ab, left_on="cod_prod", right_on="key_prod", how="left")
```

```python
#tiendas
merged_sem = merged_sem.merge(tda, left_on="Sucursal", right_on="key_tda", how="left")
```

ELIMINAR los siguientes puntos de venta

```python
#eliminar los siguientes puntos de venta
condition =(merged_sem['Sucursal'] == "E3") | \
           (merged_sem['Sucursal'] == "E5") | \
           (merged_sem['Sucursal'] == "F2") | \
           (merged_sem['Sucursal'] == "GB") | \
           (merged_sem['Sucursal'] == "BOD") | \
           (merged_sem['Sucursal'] == "CUM") | \
           (merged_sem['Sucursal'] == "LOT") | \
           (merged_sem['Sucursal'] == "MER") | \
           (merged_sem['Sucursal'] == "PAB") | \
           (merged_sem['Sucursal'] == "CUL")
# Delete rows based on the condition
merged_sem = merged_sem[~condition]
```

#Conversiones: vta/invt

Procesamiento a Inventarios

```python
#piezas a cajas
merged_stocks["pzaTOcjas"]=merged_stocks["Exist.Pzas"]/merged_stocks["Unidad de caja"]
merged_stocks["inv cajas TT"]=merged_stocks["pzaTOcjas"]+merged_stocks["Exist.Cjas"]

#seleccionar columnas por index para dataset final
merged_stocks= merged_stocks.iloc[:,[8,10,13]]
```

Procesamiento a Ventas

```python
# piezas convertidas a cajas
merged_sem["pzasTOcjas"] = merged_sem["Piezas"] / merged_sem["Factor_Caja"]
merged_sem["cjas TT"] = merged_sem["pzasTOcjas"] + merged_sem["Cajas"]

#seleccionar columnas por index para dataset final
#merged_sem= merged_sem.iloc[:,[5,8,10,13]]

# crear columna sin impuesto al valor de precios

merged_sem["$_s/iva"] = merged_sem['Total Importe Bruto'] / 1.16
```

```python
merged_sem.tail()
```

#Pivot Table

```python
#concatenar
sos=pd.concat([merged_sem, merged_stocks])

#tabla dinamica
pivot_table=pd.pivot_table(data=sos,
                           index=["key_tda", "AB actual"],
                           values=["cjas TT", "Total Importe Bruto", "inv cajas TT", "$_s/iva"],
                           aggfunc=np.sum,
                           fill_value=0)

# Assuming you want the current column order
column_order = ["key_tda", "AB actual", "cjas TT", "$_s/iva", "inv cajas TT", "Total Importe Bruto"]

pivot_table.reset_index().to_excel('sos.xlsx',sheet_name = 'Sheet1',index=False, columns= column_order )
```
