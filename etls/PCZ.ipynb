```python
z = pd.read_excel("/content/drive/MyDrive/Sell Out/PCZ/Semanal_Tissue/Vtas/2025/17_Z.xls")
mty=pd.read_excel("/content/drive/MyDrive/Sell Out/PCZ/Semanal_Tissue/Vtas/2025/17_MTY.xls")

#z = pd.read_excel("/content/drive/MyDrive/Sell Out/PCZ/Mensual_Tissue/Vta/2025/25.03_Z.xls")
#mty=pd.read_excel("/content/drive/MyDrive/Sell Out/PCZ/Mensual_Tissue/Vta/2025/25.03_MTY.xls")

sem= pd.concat([z,mty])

sem= sem.replace(r'\s+',' ', regex=True) # espacios en blanco

sem["ARTIC"] = sem["ARTIC"].astype(float)

sem.info()
```

select columns by index position


```python
#sem = sem.iloc[:, [0,6,7,8,9,14,15]]
sem = sem.iloc[:, [0,6,7,12,13,17,18]]
sem.head()
```

Transformar el tipo de dato de algunas columnas

```python
sem['F_MOVTO'] = sem['F_MOVTO'].astype('datetime64[ns]')
```

Seleccionar el rango de fechas

```python
#rango de fechas
start_date = datetime.strptime('04/21/2025', '%m/%d/%Y')
end_date = datetime.strptime('04/27/2025', '%m/%d/%Y')

#filtrar el rango de fechas dadas
sem = sem[(sem['F_MOVTO'] >= start_date) & (sem['F_MOVTO'] <= end_date)]
```

valores perdidos

```python
print("Missing values in semanal:")
print(sem.isnull().sum())
print("")
sem.shape
```

Sustituit valores negativos

```python
sem['CAJAS'] = np.where(sem['CAJAS'] < 0, 0, sem['CAJAS'])
sem['SUBTOT'] = np.where(sem['SUBTOT'] < 0, 0, sem['SUBTOT'])
```

seleccionar valores donde ambas columnas son mayores a 0 (vtas y $)

```python
sem = sem[(sem['CAJAS']>0) & (sem['SUBTOT']>0)]
```

Renombrar columnas

```python
sem = sem.rename(columns={"CVE_ALM": "determinante",
                          'ARTIC': 'sku',
                          'ARTICULO': "descripcion"}
                 )
```

eliminar todo espacio en blanco una vez obtenido el dataframe final

```python
sem = sem.drop ('F_MOVTO', axis = 1) #eliminar fecha del dataset
```

ELIMINAR los siguientes centros de venta

```python
#eliminar los siguientes puntos de venta
condition =(sem['determinante'] == 603) | \
           (sem['determinante'] == 251) | \
           (sem['determinante'] == 902) | \
           (sem['determinante'] == 602) | \
           (sem['determinante'] == 500) | \
           (sem['determinante'] == 284) | \
           (sem['determinante'] == 862) | \
           (sem['determinante'] == 2) | \
           (sem['determinante'] == 922) | \
           (sem['determinante'] == 792) | \
           (sem['determinante'] == 942) | \
           (sem['determinante'] == 842) | \
           (sem['determinante'] == 822)
# Delete rows based on the condition
sem = sem[~condition]
```

REMPLAZAR los siguientes valores

```python
sem["determinante"] = sem["determinante"].replace([832], 831)
sem["determinante"] = sem["determinante"].replace([827], 487)
```

ELIMINAR los siguientes productos

```python
#eliminar los siguientes puntos de venta
cond =(sem['sku'] ==900001)

# Delete rows based on the condition
sem = sem[~cond]
```
