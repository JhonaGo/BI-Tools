# Regresiones multivariables

```python
#libreria
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#omitir notacion cientifica
pd.options.display.float_format = '{:.2f}'.format
```

```python
df=pd.read_excel("/content/drive/MyDrive/Sell Out/M_DECASA/analisis decasa.xlsx", sheet_name="LM")
```

```python
df['Categoria'] = df['Categoria'].str.replace(' ', '_')  #remplazar nombres
df['Categoria'] = df['Categoria'].str.strip() #eliminar espacios en blanco en toda posicion
```

```python
# renombrar las columnas
df= df.rename(columns={'Suma de VOL': "Vol"})
df= df.rename(columns={'Suma de GASTO CLIENTE': "GastClte"})
df= df.rename(columns={'Suma de RAPPEL': "Rappel"})
df= df.rename(columns={'Suma de COSTO SERVIR': "CostServ"})
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['Rappel'] = df['Rappel'].fillna(0)
```

```python
df.Categoria.unique()
```

Cuidado Infantil

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df1 = df[df['Categoria'] == 'CUIDADO_INFANTIL']

df1 = df1.iloc[:, [2, 3, 4, 5]]
```

```python
#Setting the value for X and Y
x = df1[["GastClte","Rappel", "CostServ"]]
y = df1['Vol']
```

```python
#Splitting the dataset

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
```

```python
#Fitting the Multiple Linear Regression model

mlr = LinearRegression()
mlr.fit(x_train, y_train)
```

```python
#Intercept and Coefficient

print("Intercept: ", mlr.intercept_)
print("Coefficients:")
list(zip(x, mlr.coef_))
```

```python
#Prediction of test set
y_pred_mlr= mlr.predict(x_test)

#Predicted values
print("Prediction for test set: {}".format(y_pred_mlr))
```

```python
#Actual value and the predicted value
mlr_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred_mlr})
mlr_diff.head()
```

```python
#Model Evaluation
from sklearn import metrics
meanAbErr = metrics.mean_absolute_error(y_test, y_pred_mlr)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_mlr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_mlr))

print('R squared: {:.2f}'.format(mlr.score(x,y)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```
