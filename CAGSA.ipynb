### Dataset

```python
#omitir notacion cientifica
pd.options.display.float_format = '{:.2f}'.format
```

Carga del archivo: hay que cambiar la version de excel desde el archivo de origen

```python
sem = pd.read_excel("/content/drive/MyDrive/Sell Out/CAGSA/Semanal/2023/25_10.xlsx",
                    engine='openpyxl') # Explicitly use the openpyxl engine
sem.head()
```

Ajustes en el formato

Seleccion de columnas de interes

```python
sem.columns
```

```python
#select columns by index position
sem = sem.iloc[:, [2,3,4,5,6,7]]
```

```python
sem.info()
```

Valores perdidos en el dataset

```python
print("Missing values distribution: ")
print(sem.isnull().mean())
print("")
sem.shape
```

Sustituir valores negativos

```python
sem['Venta'] = np.where(sem['Venta'] < 0, 0, sem['Venta'])
sem['Importe'] = np.where(sem['Importe'] < 0, 0, sem['Importe'])
sem['Existencia *'] = np.where(sem['Existencia *'] < 0, 0, sem['Existencia *'])
```

Data frame final

```python
sem = sem.replace(r'\s+', ' ', regex=True) # espacios en blanco

sem=sem[["Sucursal", "Código", "Descripción", "Venta", "Importe", "Existencia *"]]  #reordenamiento de columnas

sem['Sucursal'] = sem['Sucursal'].str.lower()  #columna de caracteres a minisculas
sem.head()
```

### Merged catalogos

```python
#productos
merged_sem = sem.merge(ab, left_on="Código", right_on="CLAVE", how="left")
merged_sem.head()
```

```python
# tiendas
merged_sem = merged_sem.merge(tda, left_on="Sucursal", right_on="Nombre de Tienda", how="left")
merged_sem.head()
```

Observar valores faltantes en el dataset

```python
print(merged_sem.isnull().sum())
```

```python
nas= merged_sem[merged_sem.isna().any(axis=1)]  #revisar registros con datos faltantes

nas.to_excel('nas.xlsx',sheet_name = 'Sheet1',index=False)

nas.tail()

merged_sem.dropna(inplace=True) #eliminar registros con datos faltantes
```

### Dividir las bases

```python
#inventarios
invt=merged_sem[["id", "AB actual", "Existencia *"]]
invt.head()
```

```python
#ventas
vtas=merged_sem[["id", "AB actual", "Venta", "Importe"]]

vtas = vtas[(vtas['Venta']>0) & (vtas['Importe']>0)]

vtas["importe s/iva"] = vtas["Importe"] / 1.16

vtas.head()
```

### Pivot Table

```python
#unir los dos datasets de forma vertical
sos = pd.concat([vtas, invt])

#tabla dinamica
pivot_table=pd.pivot_table(data=sos,
                           index=["id", "AB actual"],
                           values=["Venta", "importe s/iva", "Existencia *", "Importe"],
                           aggfunc=np.sum,
                           fill_value=0)

# Assuming you want the current column order
column_order = ["id", "AB actual", "Venta", "Importe", "Existencia *", "importe s/iva"]

#exportar
pivot_table.reset_index().to_excel('sos.xlsx',sheet_name = 'Sheet1',index=False, columns= column_order)
```
